```python
#Importing the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sea
import pandas as pd
```

```python
#Loading the dataset
sales_data=pd.read_csv('C:\\Users\\haral\\OneDrive\\Documents\\shruti\\My Projects\\Amazon 2_Raw.csv')
```

```python
#Printing the head of the dataset
sales_data.head()
```

```python
#cleaning the dataset
sales_data.info()
print("Duplicate values in the dataset are:", sales_data.duplicated().sum())
print("Null values in the dataset are:\n ", sales_data.isnull().sum())
```

```python
#Showing the distribution of sales using a histogram 
plt.figure(figsize=(10,6))
plt.hist(sales_data['Sales'], bins=30, color='violet' , edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales') 
plt.ylabel('Frequency')
plt.show()
plt.savefig('sales_distribution.png')
```

Interpretation- The histogram shows that the sales data is highly positively skewed. Most of the sales values are near/around in the lower range, meaning that a large number of transactions have relatively small sales amounts. However, there are a few transactions with very high sales values, which create a long tail on the right side of the graph. These high-value points can be considered as outliers or exceptional sales. This indicates that while regular sales are generally low, occasional large transactions significantly impact the overall sales distribution. 

```python
#Showing the profit distribution using a KDE plot
sea.kdeplot(sales_data['Profit'], shade=True, color='blue')
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Density')
plt.show()
plt.savefig('profit_distribution.png')
```

Interpretation-The profit distribution is going from low to high and again low,so we cam say its slightly positively skewed.The KDE plot shows that the profit distribution is highly around zero, indicating that most transactions generate very less profit or are close to break-even. The sharp peak near zero suggests that a large number of observations fall in this range. Additionally, the distribution appears to be slightly right-skewed, as there are some higher profit values extending towards the right side. There are also values on the left side (negative side), indicating that some transactions result in losses. Overall, the data shows that while most profits are low, there are a few higher profit and loss values, indicating variability in profitability.

```python
#Showing the boxplot for sales  
plt.figure(figsize=(16,10))
sea.boxplot(y=sales_data['Sales']) 
plt.title('Boxplot for Sales')
plt.show()
plt.savefig('sales_boxplot.png')
```

Interpretation-The boxplot shows that the majority of sales values are around in a low range, as indicated by the compressed box near the bottom. The median sales value is relatively low, suggesting that most transactions involve smaller sales amounts. However, there are a large number of points above the upper whisker, which represent outliers. These outliers indicate that there are several transactions with unusually high sales values.
The presence of many high-value outliers suggests that the data is highly right-skewed, where a few large transactions significantly impact the overall distribution. To improve visibility, a log scale can be used, which helps in better understanding the spread of the data by reducing the effect of extreme values.

```python
#Showing the boxplot for sales with log scale
plt.figure(figsize=(16,10))
sea.boxplot(y=sales_data['Sales']) 
plt.yscale('log')
plt.title('Boxplot for Sales with Log Scale')
plt.ylabel('Sales (Log Scale)')
plt.show()
```

```python
#Comparing sales across different categories using a boxplot
plt.figure(figsize=(16,10))
sea.boxplot(x=sales_data['Category'], y=sales_data['Sales'])
plt.title('Sales across Categories')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
plt.savefig('sales_categories_boxplot.png')
```

Interpretation-The boxplot shows the distribution of sales across different product categories. It is observed that categories like Copiers, Machines,Tables,Chairs,Phones amd bookcases have higher median sales and a wider spread, indicating that these categories generate higher and more variable sales. In contrast, categories such as Labels,Art,Fasteners, and Supplies have lower median sales and a narrower spread, suggesting consistently low sales values.
Additionally, there are numerous outliers present in almost all categories, especially in categories like Phones, Binders, and Tables, indicating occasional high-value transactions. The variation in box sizes across categories shows that some categories have more variability in sales than others. Overall, the data suggests that while most categories have moderate to low sales, a few categories contribute significantly to higher sales values.

```python
#Comparing profit with sales using a scatterplot
plt.figure(figsize=(10,6))
sea.scatterplot(x=sales_data['Profit'], y=sales_data['Sales'])
plt.title('Profit vs Sales')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.show()
plt.savefig('profit_vs_sales.png')
```

Interpretation-The scatter plot shows the relationship between sales and profit. It can be observed that there is a positive relationship between sales and profit, meaning that as sales increase, profit also tends to increase. Most of the data points are clustered near the lower range of both sales and profit, indicating that the majority of transactions involve smaller sales with lower profit.<br>
There are also some points with negative profit, showing that certain sales result in losses even when sales values are present. Additionally, a few points are far from the main cluster, representing outliers with very high sales and profit values. Overall, the plot suggests that higher sales generally lead to higher profit, but not always, as some transactions still result in losses.<br>
The relationship between sales and profit is positive but not perfectly linear, indicating other factors also affect profit.
